Equation for calculation of address of $i^{th}$ element of an array is given below:

address of $i^{th}$ element = base address + ($i \times$ no of bytes to store one element)

If we have an integer array,

address of $i^{th}$ element = base address + $(i \times 4)$

# 2. Disadvantages of Arrays

1. Fixed size
    - In C/C++/Java/Python languages, arrays are fixed size. Most of the time, we declare very large size arrays which are not fully utilized $\implies$ Memory wastage
2. Homogeneous
    - Only elements of same type can be stored $\implies$ Lack of flexibility

# 3. Referential Arrays

```python
# Normal Arrays and Referential Arrays

from IPython import display
display.Image("data/images/DSA_04_Arrays-03-Referential-Arrays.jpg")
```

- In Normal arrays, elements of same type are stored in contiguous memory locations. Normal array is an array of values. Example: Arrays in C/C++/Java/Python

- In the case of Referential arrays, elements of different types are stored in random memory locations (no need to be contiguous). Referential array is an array of references (memory addresses) to these elements. Thus, we can create a Heterogeneous array of elements of different types while operating a Homogeneous array of memory addresses of these elements. Example: Python Lists.

Advantages:
- Can create Heterogeneous arrays
- Coding becomes easier

Disadvantages:
- Need extra memory
- Speed of access of elements $↓$

**Note**

Reduction in speed of access is not a big disadvantage since it is very small and will not affect most of the applications (except if you are working on a critical project for NASA where even timings in nanoseconds should be considered).
