# 2. Time Complexity

## 1. Techniques to measure Time Efficiency
We want to
- Evaluate an algorithm
- Evaluate the scalability
- Evaluate in terms of input size

Three techniques for Time Efficiency evaluation are
- Measuring execution time
- Counting operations involved
- Abstract notion of order of growth

**Note**

Different inputs change how the program runs.

Consider the following function:

```
def search_for_element(L, e):
    for element in L:
        if element == e:
            return True
    
    return False
```

- When e is 1st element $\implies$ best case
- When e is somewhere in the middle $\implies$ average case
- When e is either last element or not in the list $\implies$ worst case

We design our algorithms focusing on the worst case scenario.

### 1. Measuring execution time

````python
# Measuring execution time
import time

start = time.time()

for i in range(1, 26):
    print(f"i = {i}")

print(f"Execution time = {time.time()-start}")
````

Pros and cons of this approach are given below:

| Characteristics of the approach | Pro / Con |
| :------- | :-------- |
| Different time for different algorithm | ✅ |
| Time varies if implementation changes (eg: use of while loop instead of for loop will change execution time) | ❌ |
| Different machines $\implies$ different time | ❌ |
| Do not work for extremely small input | ❌ |
| Time varies for different inputs, but cannot extract the relationship between time and input | ❌ |

Hence, this technique is not recommended.

### 3. Orders of Growth

#### 1. Goals:

- We want to evaluate program's efficiency when **input is very big**
- We want to express the **growth of program's run time** as input size grows $\implies$ we want to plot the graph between time and input size
- We want to put an **upper bound** on growth - as tight as possible $\implies$ focus on worst case scenario
- Do not need to be precise $\implies$ **order of not, exact growth**
- We will give preference to **largest factors in run time** (which section of program will take the longest to run?)

Thus, we want tight upper bound on growth, as a function of input size, in the worst case scenario.

#### 2. Measuring Order of Growth: Big O Notation

Big O Notation measures an **upper bound on the asymptotic growth**, often called Order of Growth.

Big O is used to describe the worst case scenario:
- worst case scenarios occur often and is the bottleneck when a program runs
- express rate of growth of program relative to the input size
- evaluate algorithm, not machine or implementation

Steps to find Order of Growth:
- Find no of operations as a function of input size
- In order to find the worst case asymptotic complexity,
    - Ignore additive constants
    - Ignore multiplicative constants

Consider below function:

```
def get_factorial(n):
    answer = 1
    while n > 1:
        answer *= n
        n -= 1
    return answer
```

In the above example, we have
- `answer = 1` $\implies$ 1 operation
- `n > 1` $\implies$ 1 operation
- `answer *= n` $\implies$ 2 operations
- `n -= 1` $\implies$ 2 operations
- Total no of operations = $1 + (1+2+2) \times n$ = $1 + 5n$ operations
- After ignoring additive and multiplicative constants, Order of Growth $O(n) = n$ $\implies$ Linear time complexity

#### 3. Examples

1. No of operations = $n^2 + 2n + 2$
    - $n^2$ $\implies$ nested loop
    - $2n$ $\implies$ loop with 2 operations inside
    - $2$ $\implies$ 2 operations outside all loops
    - Time Complexity, $O(n) = n^2$ $\implies$ Quadratic time complexity

2. No of operations = $n^2 + 100000n + 3^{1000}$
    - $n^2$ $\implies$ nested loop
    - $100000n$ $\implies$ loop with 100000 operations inside
    - $3^{1000}$ $\implies$ $3^{1000}$ operations outside all loops
    - Time Complexity, $O(n) = n^2$ $\implies$ Quadratic time complexity

3. No of operations = $\log (n) + n + 4$
    - $\log (n)$ $\implies$ divide-and-conquer type strategies like Binary Search
    - $n$ $\implies$ loop with 1 operation inside
    - $4$ $\implies$ 4 operations outside all loops
    - Time Complexity, $O(n) = n$ $\implies$ Linear time complexity

4. No of operations = $0.0001n\log (n) + 300n$
    - $0.0001n\log (n)$ $\implies$ Merge Sort, Heap Sort etc.
    - $300n$ $\implies$ loop with 300 operations inside
    - Time Complexity, $O(n) = n\log (n)$
        
        **Note**
        
        This complexity is often referred to as log-linear or quasi-linear because it combines characteristics of both logarithmic and linear growth. However, the standard Big O term for this complexity is simply $O(n) = n\log (n)$.

5. No of operations = $2n^{30} + 3^n$
    - Time Complexity, $O(n) = 3^n$ $\implies$ Exponential time complexity

        **Note**

        Examples of problems that might involve exponential complexity:
        - *Brute-force search*: This naive search approach tries every possible combination to find a solution. Its complexity can be exponential for some problems.
        - *Recursion without a base case*: If a recursive function doesn't have a base case to stop the recursion, it can lead to an infinite loop and exponential growth in the number of function calls.

#### 4. Types of Orders of Growth

Order of Growth of a program can be anyone out of the below:
1. Constant
    - Eg: Accessing an element at a specific index of an array $\implies$ Do not depend on the length of the array
2. Logarithmic
    - Eg: Binary Search
3. Linear
    - Eg: Traversal algorithms (Array Traversal, Linked List Traversal), Linear Search
4. $n \log (n)$ $\implies$ also called as log-linear or quasi-linear
    - Merge Sort, Heap Sort
5. Quadratic
    - Eg: Nested loops, Some sorting algorithms (Bubble Sort, Selection Sort, Insertion Sort), Brute-Force Search algorithms
6. Exponential

# 3. Time Complexity: Challenges

### 1. Problem 1

````python
L = [1, 2, 3, 4, 5]

sum = 0
for i in L:
    sum += i
print(sum)

product = 1
for i in L:
    product *= i
print(product)
````

Solution:

For 1st for loop:
- `sum = 0` $\implies$ 1 operation
- `i in L` $\implies$ 1 operation
- `sum += i` $\implies$ 2 operations
- `print(sum)` $\implies$ 1 operation
- Loop has $n$ iterations
- No of operations = $1 + (1 + 2) \times n + 1$ = $2 + 3n$ operations
- After ignoring additive and multiplicative constants, $O(n) = n$

For 2nd for loop:
- `product = 1` $\implies$ 1 operation
- `i in L` $\implies$ 1 operation
- `product *= i` $\implies$ 2 operations
- `print(product)` $\implies$ 1 operation
- Loop has $n$ iterations
- No of operations = $1 + (1 + 2) \times n + 1$ = $2 + 3n$ operations
- After ignoring additive and multiplicative constants, $O(n) = n$

Applying Law of Addition for Big O,

$O(f(n)) + O(g(n)) = O(f(n) + g(n)) \implies$ Combined $O(n) = O(n) + O(n) = O(n + n) = O(2n) = O(n)$ $\implies$ Linear time complexity

### 2. Problem 2

````python
L = [1, 2, 3, 4, 5]

for i in L:
    for j in L:
        print(i, j)
````

Solution:

For outer for loop:
- `i in L` $\implies$ 1 operation
- We can consider the inner for loop to be taking a constant $c$ no of operations for each iteration of outer for loop
- Loop has $n$ iterations
- No of operations = $(1 + c) \times n$ operations
- After ignoring additive and multiplicative constants, $O(n) = n$

For inner for loop:
- `j in L` $\implies$ 1 operation
- `print(i, j)` $\implies$ 1 operation
- Loop has $n$ iterations
- No of operations = $(1 + 1) \times n$ = $2n$ operations
- After ignoring additive and multiplicative constants, $O(n) = n$

Applying Law of Multiplication for Big O,

$O(f(n)) \times O(g(n)) = O(f(n) \times g(n)) \implies$ Combined $O(n) = O(n) \times O(n) = O(n \times n) = O(n^2)$ $\implies$ Quadratic time complexity

### 3. Problem 3

````python
def int_to_str(i):
    digits = "0123456789"
    if i == 0:
        return "0"
    result = ""
    while i > 0:
        result = digits[i % 10] + result
        i //= 10

    return result
````

Solution:

For worst case scenario, `i` is non-zero. In that case, we have:
- `digits = "0123456789"` $\implies$ 1 operation
- `i == 0` $\implies$ 1 operation
- `result = ""` $\implies$ 1 operation
- `i > 0` $\implies$ 1 operation
- `result = digits[i % 10] + result` $\implies$ 4 operations
- `i //= 10` $\implies$ 2 operations
- Since `while` loop is dividing `i` by `10` on each iteration, no of iterations of `while` loop is logarithmic
- No of operations = $1 + 1 + 1 + (1 + 4 + 2) \times \log (n)$ = $3 + 7 \log (n)$ operations
- After ignoring additive and multiplicative constants, $O(n) = \log (n)$ $\implies$ Logarithmic time complexity

### 4. Problem 4

````python
n = 1000;
int i, j, k = 0;

for (i=n/2; i<=n; i++) {
    for (j=2; j<=n; j = j*2) {
        k = k + n/2;
    }
}
````

Solution:

- `int i, j, k = 0;` $\implies$ 3 operations
- For outer for loop:
    - Initialization `i=n/2;` $\implies$ 1 operation
    - `i<=n;` $\implies$ 1 operation
    - `i++` $\implies$ 2 operations
    - Outer for loop will be having $\frac {n}{2}$ iterations
    - We can consider the inner for loop to be taking a constant $c$ no of operations for each iteration of outer for loop
    - No of operations = $3 + 1 + (1 + 2 + c) \times \frac {n}{2}$ = $4 + (3 + c) \times \frac {n}{2}$ operations
    - After ignoring additive and multiplicative constants, $O(n) = n$

- For inner for loop:
    - Initialization `j=2;` $\implies$ 1 operation
    - `j<=n;` $\implies$ 1 operation
    - `j = j*2;` $\implies$ 2 operations
    - `k = k + n/2;` $\implies$ 3 operations
    - Since inner `for` loop is multiplying `j` by `2` on each iteration, no of iterations of `for` loop is logarithmic
    - No of operations = $1 + (1 + 2 + 3) \times \log (n)$ = $1 + 6 \log (n)$ operations
    - After ignoring additive and multiplicative constants, $O(n) = \log (n)$

Applying Law of Multiplication for Big O,

$O(f(n)) \times O(g(n)) = O(f(n) \times g(n)) \implies$ Combined $O(n) = O(n) \times O(n) = O(n \times \log (n))$ $\implies$ Log-linear or Quasi-linear time complexity

### 5. Problem 5

````python
L = [1, 2, 3, 4, 5]

for i in range(0, len(L)):
    for j in range(i+1, len(L)):
        print(L[i], L[j])
````

Solution:

- `i in range(0, len(L))` $\implies$ 1 operation
- Outer for loop will be having $n$ iterations
- `j in range(i+1, len(L))` $\implies$ 1 operation
- `print(L[i], L[j])` $\implies$ 1 operation
- No of iterations of the inner for loop = $n - i$
- No of operations = $1 + (n - 0) \times 2 + 1 + (n - 1) \times 2 + 1 + (n - 2) \times 2 + ...$ for $n$ terms = $n + 2 \times (1 + 2 + 3 + ... + n)$ = $n + 2 \times \frac {n \times (n + 1)}{2}$ = $n + n^2 + n$ = $n^2 + 2n$ operations
- After ignoring additive and multiplicative constants, $O(n) = n^2$ $\implies$ Quadratic time complexity

### 6. Problem 6

````python
A = [1, 2, 3, 4]
B = [2, 3, 4, 5, 6]

for i in A:
    for j in B:
        if i < j:
            print(i, j)
````

Solution:

Assuming size of array A as $a$ and size of array B as $b$,
- `i in A` $\implies$ 1 operation
- Outer for loop will be having $a$ iterations
- `j in B` $\implies$ 1 operation
- `i < j` $\implies$ 1 operation
- `print(i, j)` $\implies$ 1 operation
- Inner for loop will be having $b$ iterations
- In the worst case scenario, `if i < j` will be returning `True` in all iterations and `print(i, j)` will be executed every time
- In that case, no of operations of the inner for loop = $(1 + 1 + 1) \times b = 3 \times b$
- Total no of operations = $a \times 3 \times b = 3 \times ab$ operations
- After ignoring additive and multiplicative constants, $O(n) = ab$ $\implies$ Quadratic time complexity

### 7. Problem 7

````python
A = [1, 2, 3, 4]
B = [2, 3, 4, 5, 6]

for i in A:
    for j in B:
        for k in range(100000):
            print(i, j)
````

Solution:

Assuming size of array A as $a$ and size of array B as $b$,
- `i in A` $\implies$ 1 operation
- Outermost for loop will be having $a$ iterations
- `j in B` $\implies$ 1 operation
- Inner for loop will be having $b$ iterations
- `k in range(100000)` $\implies$ 1 operation
- `print(i, j)` $\implies$ 1 operation
- Innermost for loop will be having 100000 iterations
- No of operations of the innermost for loop = $(1 + 1) \times 10000 = 20000$
- No of operations of the inner for loop = $(1 + 20000) \times b = 20001 \times b$
- Total no of operations = $(1 + 20001 \times b) \times a = a + 20001 \times ab$ operations
- After ignoring additive and multiplicative constants, $O(n) = ab$ $\implies$ Quadratic time complexity

### 8. Problem 8

````python
L = [1, 2, 3, 4, 5]

for i in range(0, len(L)//2):
    other = len(L) - i - 1
    temp = L[i]
    L[i] = L[other]
    L[other] = temp
````

Solution:

- `i in range(0, len(L)//2)` $\implies$ 1 operation
- `other = len(L) - i - 1` $\implies$ 4 operations
- `temp = L[i]` $\implies$ 2 operations
- `L[i] = L[other]` $\implies$ 2 operations
- `L[other] = temp` $\implies$ 1 operation
- The for loop will be having $\frac {n}{2}$ iterations
- Total no of operations = $(1 + 4 + 2 + 2 + 1) \times \frac {n}{2} = 5n$ operations
- After ignoring additive and multiplicative constants, $O(n) = n$ $\implies$ Linear time complexity

### 9. Problem 9

````python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
````

Solution:

In the case of loops, we check no of iterations. In the case of Recursion, we check no of function calls. We can consider each function call to be taking a constant $c$ no of operations. In that case, we have

- Total no of operations = $c \times n$ operations
- After ignoring additive and multiplicative constants, $O(n) = n$ $\implies$ Linear time complexity
