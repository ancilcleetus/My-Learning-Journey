# 1. What is Deep Learning

Deep Learning is a subset of Machine Learning. Deep Learning uses Artificial Neural Networks that are capable of Representation Learning.

In Traditional Machine Learning approaches, features are input to ML Algorithms manually by humans (called Feature Extraction or Feature Engineering) whereas in Deep Learning, Neural Networks learn the features automatically.

In Deep Learning, Neural Networks have many hidden layers between input and output layers. These hidden layers extract features from lower level to higher level. For example, consider CNNs used in Computer Vision. Here, initial hidden layers extract lower level features such as edges followed by simple shapes while subsequent hidden layers extract higher level features such as more complex shapes. The final hidden layer identifies things relevant to that particular problem - say digits, letters, animals, faces etc.
