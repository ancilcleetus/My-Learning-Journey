# 1. What is Deep Learning

Deep Learning is a subset of Machine Learning. Deep Learning uses Artificial Neural Networks that are capable of Representation Learning.

In Traditional Machine Learning approaches, features are input to ML Algorithms manually by humans (called Feature Extraction or Feature Engineering) whereas in Deep Learning, Neural Networks learn the features automatically.

```python
# Deep Learning as subset of ML

from IPython import display
display.Image("data/images/DL_01_Intro-01-DL-subset-of-ML.jpg")
```

In Deep Learning, Neural Networks have many hidden layers between input and output layers. These hidden layers extract features from lower level to higher level. For example, consider CNNs used in Computer Vision. Here, initial hidden layers extract lower level features such as edges followed by simple shapes while subsequent hidden layers extract higher level features such as more complex shapes. The final hidden layer identifies things relevant to that particular problem - say digits, letters, animals, faces etc.

# 2. Deep Learning Vs Machine Learning (Practical Differences)

## 1. Data Dependency

For small datasets, ML performs same as or even better than DL. With more data, there is no improvement in performance of ML (performance stagnates). DL is data hungry and its performance increases linearly with more data.

```python
# Data Dependency

from IPython import display
display.Image("data/images/DL_01_Intro-03-DL-Performance.jpg")
```

## 2. Hardware Dependency

ML Algorithms can run seamlessly in CPUs. DL Algorithms need to do very complex matrix multiplications (matrices with thousands or even millions of values) and hence need powerful GPUs. DL algorithms are really slow in CPUs. Hence, DL requires costly hardware in comparison to ML.

## 3. Training Time

DL models are very complex and require huge training times (days to weeks to even months). ML models are relatively less complex and take very less time to train (within several minutes or hours).

Inference / Prediction time is very fast for DL models whereas it varies depending on ML model used. Prediction is slow for some ML algorithms (eg: KNN).

## 4. Feature Selection

DL models extract relevant features automatically from data (Representation Learning). Initial hidden layers extract lower level features while mid-level and higher level features are extracted in subsequent hidden layers. ML models need the features to be manually extracted (Feature Engineering).

```python
# Feature Selection

from IPython import display
display.Image("data/images/DL_01_Intro-04-Feature-Selection.png")
```

## 5. Interpretability / Explainability

ML models have very high interpretability. For example,
1. in the case of Logistic Regression, we can check the weights for each feature and give explanation for a particular prediction.
2. in the case of Decision Trees, we have a flow chart based on which we can explain a particular prediction.

The hidden layers in a DL model function like a black box. Since features are automatically extracted by a DL model, we don't know whether those features are actually relevant for that particular problem. For example, a Dogs Vs Cats Classifier DL model will classify an image as either a Dog or Cat, but it cannot give the reason for that particular prediction. Hence, DL models have very low interpretability.

Hence in scenarios where we need to provide both the result and why we got that result, DL is not a good choice.

## DL Vs ML: Conclusion

Just as a sword cannot replace a pen knife, DL cannot replace ML. There will always be scenarios for which ML is desired - especially for small datasets and where explainability is required.

# 4. Types of Neural Networks

By arranging layers and nodes in different ways and by using different types of nodes, we can have different types of DL architectures.

1. MLPs (Multi-Layer Perceptrons)
    - Supervised ML Regression / Classification Problems where non-linear relationship needs to be captured
2. CNNs (Convolutional Neural Networks)
    - Image Processing and Video Processing Applications
3. RNNs (Recurrent Neural Networks)
    - NLP (Natural Language Processing) Applications
4. Auto-encoders
    - Compression without losing quality
5. GANs (Generative Adversarial Networks)
    - Generate images, music, stories through imagination

# 5. History of Deep Learning

1. 1958: Rosenblatt's Single-Layer Perceptron
2. 1969: Minsky-Papert -> Single-Layer Perceptron cannot learn a simple XOR function
3. 1969 onwards: First AI Winter
4. 1980's: The Rise
5. 1986: Backpropagation Algorithm by Geoffrey Hinton
6. 1989: Universal Approximation Theorem by Cybenko, CNN by Yann LeCun
7. 1990's: Realization that current tools & techniques, research, hardware and datasets cannot do much + Rise of Traditional ML Algorithms like SVM that gave good results => Second AI Winter
8. 2006: Algorithms for training Deep Belief Networks by Geoffrey Hinton
9. 2012: ImageNet Challenge (AlexNet trained using GPU -> Reduced Classification Error from 26 % to 16 %) => The Final Rise of Deep Learning
